## Bar plot in seaborn  python 

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd
```

```python
varb =  sns.load_dataset("penguins")
varb
```

```python

plt.figure(figsize=(4,3))  # Adjust the size (width=6, height=4)

sns.barplot(x=varb.island,y=varb.bill_length_mm,color="y")

# parameter of barplot  # in this data male and female are mix in next we will divide 
# we can also change order as show torgerson is 1st , biscoeislan 2nd  and dream 3rd  but i want order at first then in next we change 

plt.title("penguins")

plt.show()
```

```python
# second method to show graph 

plt.figure(figsize=(4,3))  # Adjust the size (width=6, height=4)

order_1 = ["Dream","Torgersen","Biscoe"]

sns.barplot(x="island",y="bill_length_mm",data=varb,hue="sex",order=order_1,saturation=0.3)


# direct passing verb as data at once time not repeat again and agian 
# parameter of barplot  # for partition we use hue parameter we show sub partion in the graph
# saturation parameter used for bright or dark the color  blur the graph

plt.title("penguins")
plt.show()
```

```python
# hue_order parameter used for female first then men 


plt.figure(figsize=(4,3))  # Adjust the size (width=6, height=4)

order_1 = ["Dream","Torgersen","Biscoe"]

sns.barplot(x="island",y="bill_length_mm",data=varb,hue="sex",order=order_1,hue_order=["Female","Male"],palette="BuPu",errcolor="g")

# ci=100 paramter the line show on every candle is ci its value from 0 to 100 
# palette use for changing the color there are so many list you will proviude 

plt.title("penguins")
plt.show()
```

```python
# if we want to change the grpah orientation to horizontal there is a some certain condition both x and y are numericl not a categorical 


plt.figure(figsize=(4,3))  # Adjust the size (width=6, height=4)

order_1 = ["Dream","Torgersen","Biscoe"]

sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=varb,orient="h")
# for orient parameter we will convert into vertical and horizontal 
# "errcolor", "errwidth"  used for the bar show on the candle  which is black then we change its color and weidth
# also draw a cap  on the bar "capsize" = 0.2 like this 

# alpha parameter also used for blue the grph "alpha=0.3"


plt.title("penguins")
plt.show()
```

```python
# Grid in seaborn 
sns.set(style="darkgrid") # applying dark grid 


plt.figure(figsize=(4,3))  # Adjust the size (width=6, height=4)

order_1 = ["Dream","Torgersen","Biscoe"]

sns.barplot(x="island",y="bill_length_mm",data=varb,hue="sex",order=order_1,saturation=0.3)




plt.title("penguins")
plt.show()
```

## Heatmap plot  in  Seaborn 

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
import numpy as np 
# heat map work on 2d array
# for show our data in color bar we use heatmap 
```

```python
var = np.linspace(1,10,20).reshape(4,5)# convert the data in 2d array 
var
```

```python
# Creating  heatmap graph 
plt.figure(figsize=(5,4))  # Adjust the size (width=6, height=4

sns.heatmap(var)
```

```python
# finding cmap parameter 
import matplotlib.pyplot as plt

print(plt.colormaps())  # List of all colormaps
```

```python
# parameter min or max
# changing the color parameter we use cmap 
# annot parameter  the value pass in heatmap also show in the heatmap graph 

# Creating  heatmap graph 
plt.figure(figsize=(5,4))  # Adjust the size (width=6, height=4

sns.heatmap(var,vmin=0,vmax=10,cmap="PuOr",annot=True)
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)# convert the data in 2d array 
var1
```

```python
plt.figure(figsize=(5,4))  # Adjust the size (width=6, height=4
# applying  multiple parameter then we pass a dictionary  we use "annot_kws " parameter 
# "line width" parameter add space between them also aplly "line color' 
y = {"fontsize":15,"color":"b"}
sns.heatmap(var1,vmin=0,vmax=10,cmap="PuOr",annot=True,annot_kws=y,linewidths=5,linecolor="r")
```

```python
# removing color bar using "cbar"
# also removing the label parameter using " xticlabel " and "yriclabel"
plt.figure(figsize=(5,4))  # Adjust the size (width=6, height=4

y = {"fontsize":15,"color":"b"}
sns.heatmap(var1,vmin=0,vmax=10,cmap="PuOr",annot=True,annot_kws=y,linewidths=5,cbar=False,xticklabels=False,yticklabels=False)
```

```python
# applying our own label  paramter 
plt.figure(figsize=(5,4))  # Adjust the size (width=6, height=4

y = {"fontsize":15,"color":"b"}
v= sns.heatmap(var1,vmin=0,vmax=10,cmap="PuOr",annot=True,annot_kws=y,linewidths=5,cbar=False,xticklabels=False,yticklabels=False)
v.set(xlabel="python",ylabel="wscube")
plt.show()
```

## violin plot in seabron 

```python
# similar as box and whisker plot

import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd
```

```python
vio =  sns.load_dataset("tips")
vio
```

```python

plt.figure(figsize=(5,4))  # Adjust the size (width=6, height=4

sns.violinplot(x="day",y="total_bill",data=vio,hue="time")
plt.show()
```

```python
# parameter  line width 

plt.figure(figsize=(5,4))  # Adjust the size (width=6, height=4

sns.violinplot(x="day",y="total_bill",data=vio,linewidth=3,palette="Dark2",saturation=0.4)
plt.show()
```

```python

plt.figure(figsize=(5,4))  # Adjust the size (width=6, height=4
# show total data
sns.violinplot(x="time",y="total_bill",data=vio,order=["Dinner","Lunch"],saturation=0.3,color="y",inner="stick")
# inner parameter change the inner style first he show bar then show quart paramete used in inner ("box","point","stick",quartile ,nono)
plt.show()
```

```python

plt.figure(figsize=(5,4))  # Adjust the size (width=6, height=4
# show total data
sns.violinplot(x="day",y="total_bill",data=vio,hue="sex",split=True)
plt.show()
```

```python
# for horizonta 

plt.figure(figsize=(5,4))  # Adjust the size (width=6, height=4
# show total data
sns.violinplot(x="total_bill",y="day",data=vio,hue="sex",split=True)
plt.show()
```

```python
# for single violit 

plt.figure(figsize=(5,4))  # Adjust the size (width=6, height=4
sns.violinplot(y=vio["total_bill"])
plt.show()
```

## Pair plot in seaborn 

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
```

```python
pair =  sns.load_dataset("tips")
pair
```

```python

plt.figure(figsize=(4,3))  # Adjust the size (width=5, height=4)

sns.pairplot(pair,hue="sex")
plt.show()
# this is the pair plot of all data if we want osome spectfic pair plot then we pass vars as list
```

```python

plt.figure(figsize=(4,3))  # Adjust the size (width=5, height=4)

sns.pairplot(pair,vars=["total_bill","tip"],hue="sex") # this graph show beteeen two specfic column 
plt.show()
```

```python
# chaning the hue orderabs 

plt.figure(figsize=(4,3))  # Adjust the size (width=5, height=4)

sns.pairplot(pair,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="bwr") 
plt.show()
```

```python
import seaborn as sns

# Show available built-in color palettes
print(sns.palettes.SEABORN_PALETTES.keys())
```

```python
# changing graph  using kind parameter 

plt.figure(figsize=(4,3))  # Adjust the size (width=5, height=4)

sns.pairplot(pair,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],kind="reg") 
plt.show()
```

## Strip plot in seaborn  python

```python
# strip plot basically a scatter plot 
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
```

```python
strip =  sns.load_dataset("tips")
strip
```

```python
plt.figure(figsize=(4,3))  # Adjust the size (width=5, height=4)

sns.stripplot(x="day", y="total_bill",data=strip,hue="sex",palette="pastel",linewidth=1,edgecolor="b",jitter=5)
# jitter parameter used for space between them
plt.show()
```

```python
# increase the size 
plt.figure(figsize=(4,3))  # Adjust the size (width=5, height=4)

sns.stripplot(x="day", y="total_bill",data=strip,hue="sex",palette="pastel",linewidth=1,edgecolor="y",size=8,marker=">",alpha=0.7)
plt.show()
```

```python
# single plot of strip plot
plt.figure(figsize=(4,3))  # Adjust the size (width=5, height=4)

sns.stripplot(x=strip["total_bill"])
```

## box plot in seabron 

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
```

```python
box =  sns.load_dataset("tips")
box
```

```python

plt.figure(figsize=(4,3))  # Adjust the size (width=5, height=4)

sns.set(style="whitegrid")
sns.boxplot(data=box,x="day",y="total_bill",hue="sex")
plt.show()
```

```python

# change the order 

plt.figure(figsize=(4,3))  # Adjust the size (width=5, height=4)

sns.set(style="whitegrid")
sns.boxplot(data=box,x="day",y="total_bill",hue="sex",order=["Fri","Sun","Thur","Sat"],showmeans=True,meanprops={"marker":"*","markeredgecolor":"y"}
            ,linewidth=3,palette="plasma")
plt.show()
```
