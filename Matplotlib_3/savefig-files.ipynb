## Savefig in matplotlib 

```python
# the plot we draw save externally 
import matplotlib.pyplot as plt 

x= [1,2,3,4,5]
y = [2,4,3,6,4]

plt.figure(figsize=(4,3))

plt.plot(x,y,color="r")
plt.savefig("Line123.jpg",dpi=2000,facecolor="g")
# dpi mean dot per inch  graph look very good ,  the file save by default is png but  we can also change 

plt.show()
```
